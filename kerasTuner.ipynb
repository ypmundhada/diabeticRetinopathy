```python
import pandas as pd
retina_df = pd.read_csv('/content/trainLabels.csv');
retina_df
```

```python
import os
import numpy as np
listing = os.listdir("../content/train") 
num_imgs=np.size(listing)
num_imgs
```

```python
import shutil
import sys
import os
train_dir=r'../content/train'
DR=r'../content/train/'
# if not os.path.exists(DR):
#   os.mkdir(DR)
for filename,class_name in retina_df.values:
  if(not os.path.exists(DR+str(class_name))):
    os.mkdir(DR+str(class_name))
  src = train_dir+'/'+filename+'.jpeg'
  dst = DR + str(class_name)+'/'+filename+'.jpeg'
  try:
    shutil.move(src,dst);
    print("done"+filename);
  except IOError as e:
    print('Unable to move')
```

```python
from PIL import Image
from keras.preprocessing import image
import os
import numpy as np
import pandas as pd

from subprocess import check_output
from keras.preprocessing.image import ImageDataGenerator
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
batch_size = 32
nb_classes = 5
nb_epoch = 5
nb_filters = 32
nb_pool = 2
nb_conv = 3
```

```python
train_datagen = ImageDataGenerator(
        rescale=1.0/255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        validation_split=0.2)

train_generator = train_datagen.flow_from_directory(
        './train',
        target_size=(200, 200), 
        batch_size=num_imgs,
        class_mode='categorical',
        subset='training')  

validation_generator = train_datagen.flow_from_directory(
        './train',
        target_size=(200, 200),
        batch_size=num_imgs,
        class_mode='categorical',
        subset='validation')
```

```python
train_img,train_labels = train_generator.next()
test_img,test_labels = validation_generator.next()
train_img.shape
test_img.shape
```

```python
import keras_tuner
from tensorflow import keras
from keras.optimizers import adam_v2
from keras.engine.sequential import Sequential
from keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D, Activation
img_rows, img_cols = 200, 200
```

```python
def build_model(hp):
    model = Sequential()
    model.add(Conv2D(
                filters=hp.Int('conv_1_filter', min_value=32, max_value=128, step=16), 
                kernel_size=hp.Choice('conv_1_kernel', values = [3,5]),
                activation='relu',
                input_shape=(200,200,3)))
    
    model.add(Conv2D(
                filters=hp.Int('conv_2_filter', min_value=32, max_value=64, step=8), 
                kernel_size=hp.Choice('conv_2_kernel', values = [3,5]),
                activation='relu'))
    model.add(MaxPooling2D(pool_size=(nb_pool, nb_pool)))
    
    model.add(Conv2D(
                filters=hp.Int('conv_2_filter', min_value=32, max_value=64, step=8), 
                kernel_size=hp.Choice('conv_1_kernel', values = [3,5]),
                activation='relu'))
    model.add(MaxPooling2D(pool_size=(nb_pool, nb_pool)))
    
    model.add(Flatten())
    model.add(Dense(
            units=hp.Int('dense_1_units', min_value=32, max_value=128, step=16),
            activation='relu'))
    model.add(Dense(
            units=hp.Int('dense_2_units', min_value=64, max_value=128, step=16),
            activation='relu'))
    model.add(Dense(units=5,activation='softmax'))
    model.compile(loss='categorical_crossentropy', 
                  optimizer=adam_v2.Adam(hp.Choice('learning_rate', values=[1e-2, 1e-3,1e-4])),
                  metrics=['accuracy'])
    return model
```

```python
from keras_tuner import RandomSearch
from keras_tuner.engine.hyperparameters import HyperParameters
tuner = RandomSearch(build_model,
                    objective='val_accuracy',
                    max_trials=10,
                    directory='output',
                    project_name='DR Detection'
                    )
```

```python
tuner.search(train_img,train_labels,
            epochs=5,
            validation_data=(test_img,test_labels),
            batch_size=32)
```

```python
md_best=tuner.get_best_models(num_models=1)[0]
md_best.summary()
```

```python
md_best.fit(train_img,train_labels,epochs=25,validation_data=(test_img,test_labels))
```

```python
def b_model(hp):
  model=Sequential()
  for i in range(hp.Int('num_layers',2,20)):
    model.add(Dense(units=hp.Int('units_'+str(i),
                                        min_value=32,
                                        max_value=512,
                                        step=32),
                           activation='relu'))
    model.add(Dense(5,activation='softmax'))
    model.compile(optimizer=adam_v2.Adam(hp.Choice('learning_rate', values=[1e-2, 1e-3,1e-4])),
                  loss='categorical_crossentropy', metrics=['accuracy'])
    return model;
```

```python
from keras_tuner import RandomSearch
from keras_tuner.engine.hyperparameters import HyperParameters
tuner = RandomSearch(b_model,
                    objective='val_accuracy',
                    max_trials=5,
                    directory='output',
                    project_name='DR Detection'
                    )
```
